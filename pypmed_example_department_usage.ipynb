```python
import pandas as pd 
columns=['author_first_name','author_last_name','position','department','institution']
df = pd.read_csv('sample_data/upitt_immune_v2.csv', header=None, names=columns)
df
```

```python
from pypmed import apis 
import time

all_articles = []
for year in [2022,2023]:
  for i in range(df.shape[0]):
    row = df.iloc[i]
    author_first_name = row['author_first_name'].lower()
    author_last_name = row['author_last_name'].lower()
    institution = row['institution'].lower()

    search_criteria = {'author_first_name': author_first_name, 
                      'author_last_name': author_last_name, 
                      'institution':  institution.lower(),
                      'publication_year': year
                      }

    try:
        articles = apis.query_pubmed_api(search_criteria=search_criteria)
        if articles is not None:
          all_articles.append(articles)
    except:
      pass
    
    print(f'finished row {i} author {author_first_name} {author_last_name} for year {year}, pausing 3 secs')
    time.sleep(2)
```
